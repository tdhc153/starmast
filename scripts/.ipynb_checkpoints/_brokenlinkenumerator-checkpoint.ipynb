```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Scrape https://starmast.org/ for all terms in square brackets that are NOT inside hyperlinks,
and output:
  1) non_hyperlink_bracketed_terms_dedup.csv
  2) non_hyperlink_bracketed_terms_by_page.csv

Notes:
- Respects robots.txt disallow rules (best-effort).
- Polite rate limiting.
- Deduplication + frequency counts.
- Seeded from key entry points (Full index + lists + Home/About/News) to ensure coverage.
"""

import re
import time
import csv
import sys
import queue
import urllib.parse as up
import threading
from collections import defaultdict
from pathlib import Path

import requests
from bs4 import BeautifulSoup
import urllib.robotparser as robotparser

BASE = "https://starmast.org"
SEEDS = [
    f"{BASE}/fullindex",       # Full index of materials (primary seed) – cited in chat
    f"{BASE}/studyguidelist",  # List of study guides
    f"{BASE}/fslist",          # List of factsheets
    f"{BASE}/newsindex",       # News archive
    f"{BASE}/",                # Home
    f"{BASE}/about",           # About
]

HEADERS = {
    "User-Agent": "M365Copilot-STARMastCrawler/1.0 (+https://starmast.org/; academic research)",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
}
TIMEOUT = 20
RATE_LIMIT_SECONDS = 0.5       # ~2 req/s
MAX_WORKERS = 8                # Conservative parallelism

# Regex for [bracketed text]
BRACKET_RE = re.compile(r"\[(.+?)\]")

def can_fetch(url: str, rp: robotparser.RobotFileParser) -> bool:
    try:
        return rp.can_fetch(HEADERS["User-Agent"], url)
    except Exception:
        # Fail-open (still honor base domain)
        return True

def normalize_url(url: str) -> str:
    url = url.strip()
    if not url:
        return ""
    url = up.urljoin(BASE, url)
    parsed = up.urlparse(url)
    if parsed.netloc != up.urlparse(BASE).netloc:
        return ""  # external
    # drop fragments and normalize path
    cleaned = parsed._replace(fragment="", query=parsed.query).geturl()
    return cleaned

def extract_links(soup: BeautifulSoup) -> set:
    links = set()
    for a in soup.find_all("a", href=True):
        nu = normalize_url(a["href"])
        if nu:
            links.add(nu)
    return links

def text_nodes_excluding_links(soup: BeautifulSoup):
    # Yield text from nodes that are not inside <a> tags
    for element in soup.find_all(text=True):
        parent = element.parent
        # Skip text inside scripts/styles/noscript and inside links
        if parent is None:
            continue
        if parent.name in ("script", "style", "noscript"):
            continue
        # If any ancestor is an anchor, skip
        anc = parent
        inside_link = False
        while anc is not None:
            if getattr(anc, "name", None) == "a":
                inside_link = True
                break
            anc = getattr(anc, "parent", None)
        if inside_link:
            continue
        txt = element.strip()
        if txt:
            yield txt

def fetch(url: str) -> str:
    resp = requests.get(url, headers=HEADERS, timeout=TIMEOUT)
    resp.raise_for_status()
    return resp.text

def crawl():
    # Robots
    rp = robotparser.RobotFileParser()
    try:
        rp.set_url(up.urljoin(BASE, "/robots.txt"))
        rp.read()
    except Exception:
        pass

    visited = set()
    to_visit = queue.Queue()
    for s in SEEDS:
        to_visit.put(s)

    terms_by_page = defaultdict(list)
    discovered = set()

    lock = threading.Lock()
    last_request_time = [0.0]

    def throttle():
        elapsed = time.time() - last_request_time[0]
        if elapsed < RATE_LIMIT_SECONDS:
            time.sleep(RATE_LIMIT_SECONDS - elapsed)
        last_request_time[0] = time.time()

    def worker():
        while True:
            try:
                url = to_visit.get_nowait()
            except queue.Empty:
                return
            if url in visited:
                to_visit.task_done()
                continue
            if not can_fetch(url, rp):
                to_visit.task_done()
                continue
            try:
                throttle()
                html = fetch(url)
            except Exception:
                to_visit.task_done()
                continue
            try:
                soup = BeautifulSoup(html, "lxml")

                # Collect bracketed terms outside hyperlinks
                page_terms = []
                for txt in text_nodes_excluding_links(soup):
                    for m in BRACKET_RE.finditer(txt):
                        term = m.group(1).strip()
                        if term:
                            page_terms.append(term)

                with lock:
                    visited.add(url)
                    if page_terms:
                        terms_by_page[url].extend(page_terms)

                    # Discover internal links to keep crawling
                    for nu in extract_links(soup):
                        if nu and nu not in visited and nu not in discovered:
                            discovered.add(nu)
                            to_visit.put(nu)
            finally:
                to_visit.task_done()

    threads = []
    for _ in range(MAX_WORKERS):
        t = threading.Thread(target=worker, daemon=True)
        t.start()
        threads.append(t)

    to_visit.join()
    return terms_by_page

def main():
    terms_by_page = crawl()

    # Write page-by-page occurrences
    p1 = Path("non_hyperlink_bracketed_terms_by_page.csv")
    with p1.open("w", newline="", encoding="utf-8") as f:
        w = csv.writer(f)
        w.writerow(["url", "term"])
        for url, terms in sorted(terms_by_page.items()):
            for term in terms:
                w.writerow([url, term])

    # Global counts (dedup + frequency)
    counts = defaultdict(int)
    for terms in terms_by_page.values():
        for t in terms:
            counts[t] += 1

    p2 = Path("non_hyperlink_bracketed_terms_dedup.csv")
    with p2.open("w", newline="", encoding="utf-8") as f:
        w = csv.writer(f)
        w.writerow(["term", "global_count"])
        for term in sorted(counts.keys(), key=lambda x: (x.lower(), x)):
            w.writerow([term, counts[term]])

    print("Done.\n- non_hyperlink_bracketed_terms_by_page.csv\n- non_hyperlink_bracketed_terms_dedup.csv")

if __name__ == "__main__":
    sys.exit(main())
```
